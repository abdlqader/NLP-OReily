**character RE**

```
nltk.download('words')
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

```
#let's search for words ends with ed
print([w for w in wordlist if re.search('ed$',w)])
```

```
#8 charactetrs contain j as 3rd char and t as 6th char ?
print([w for w in wordlist if re.search('^..j..t..$',w)])
```

```
#optional char ?
print([w for w in wordlist if re.search('e-?mail$',w)])
```

```
#range char
print([w for w in wordlist if re.search('^[ghi][mno][jlk][def]$',w)])
```

```
#search for decimals ?
nltk.download('treebank')
wsj = sorted(set(nltk.corpus.treebank.words()))
print([ w for w in wsj if re.search('^[0-9]+\.[0-9]+$',w)])
```

```
#search for a year
print([w for w in wsj if re.search('^[0-9]{4}$',w)])
```

```
#find vowels with 2 or more occurencs
fd = nltk.FreqDist( vs for word in wsj for vs in re.findall('[aeiou]{2,}',word))
fd.items()
```

```
#let's combine with conditionalfreqdist
nltk.download('toolbox')
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for word in rotokas_words for cv in re.findall('[ptksvr][aeiou]',word)]
cfd = nltk.ConditionalFreqDist(cvs)
cfd.tabulate()
```

```
#it will be helpful to have indx right ?
cv_word_pairs = [(cv,w) for w in rotokas_words for cv in re.findall('[ptksvr][aeiou]',w)]
cv_index = nltk.Index(cv_word_pairs)
#cv_index is a dict of all occurence for example k followed by a
print(cv_index['ka'])
print(cv_index['su'])
```

You can use a special kind of regular expression for searching across multiple words in a text (where a text is a list of tokens). For example, "\<a> \<man>" finds all instances of a man in the text. The angle brackets are used to mark token boundaries, and any whitespace between the angle brackets is ignored (behaviors that are unique to NLTK’s findall() method for texts)

```
 from nltk.corpus import gutenberg
 nltk.download('gutenberg')
 moby = nltk.Text(gutenberg.words('melville-moby_dick.txt')) 
 moby.findall('<a> (<.*>) <man>')
```

**finding the word stems(lemma) ?**

```
raw = """"DENNIS: Listen, strange women lying in ponds distributing swords is no basis for a system of government. 
Supreme executive power derives from  a mandate from the masses, not from some farcical aquatic ceremony.""" 
tokens = nltk.word_tokenize(raw)
print(tokens)
```

```
porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
print([porter.stem(t) for t in tokens])
print([lancaster.stem(t) for t in tokens])
```

The Porter and Lancaster stemmers follow their own rules for stripping affixes. Observe that the Porter stemmer correctly handles the word lying (mapping it to lie), whereas the Lancaster stemmer does not. 

```
nltk.download('wordnet')
wnl = nltk.WordNetLemmatizer()
print([wnl.lemmatize(t) for t in tokens])
```

The WordNet lemmatizer removes affixes only if the resulting word is in its dictionary. This additional checking process makes the lemmatizer slower than the stemmers just mentioned. Notice that it doesn’t handle lying, but it converts women to woman. The WordNet lemmatizer is a good choice if you want to compile the vocabulary of some texts and want a list of valid lemmas (or lexicon headwords).

```
raw =  """'When I'M a Duchess,' she said to herself, (not in a very hopeful tone though), 'I won't have any pepper in my kitchen AT ALL. Soup does very
well without--Maybe it's always pepper that makes people hot-tempered,'...""" 
#we are walking through tokenizing, let's skip the white space in the book and jump to this
print(re.split(r'\W+',raw))
```

```
print(re.findall(r'\w+|\S\w*',raw))
```

```
print(re.findall(r"\w+(?:[-']\w+)*|[-.(]+\S\w*",raw))
```

The function **nltk.regexp_tokenize()** is similar to **re.findall()** (as we’ve been using it for tokenization). However, **nltk.regexp_tokenize()** is more efficient for this task, and avoids the need for special treatment of parentheses.

```
text = 'That U.S.A. poster-print costs $12.40...'
pattern = r'''(?x)
([A-Z]\.)+ 
| \w+(-\w+)* 
| \$?\d+(\.\d+)?%?
| \.\.\. 
| [][.,;"'?():-_`]
'''
print(nltk.regexp_tokenize(text,pattern))
```
